```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importar una serie de numeros
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y,'r')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
#hitograma
plt.hist(x)
plt.show()
```

```python
# grafica de torta
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
