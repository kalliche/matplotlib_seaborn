```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
# Suponiendo que 'sex' y 'smoker' son columnas no numéricas en el DataFrame 'tips'
# Puedes excluirlas o convertirlas en valores numéricos según sea necesario
datos_numericos_tips = tips.select_dtypes(include=['number'])

# Calcular la matriz de correlación para las columnas numéricas
matriz_de_correlacion = datos_numericos_tips.corr()
matriz_de_correlacion
```

```python
sns.set_theme()
sns.heatmap(matriz_de_correlacion.corr())
plt.show()
```

```python
sns.heatmap(matriz_de_correlacion.corr(), annot=True)
plt.show()
```

```python
sns.heatmap(matriz_de_correlacion.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(matriz_de_correlacion.corr(), annot=True, cmap='coolwarm', linewidths=5)
plt.show()
```

```python
sns.heatmap(matriz_de_correlacion.corr(), annot=True, cmap='coolwarm', linewidths=5, vmin=0, vmax=1)
plt.show()
```

```python
sns.heatmap(matriz_de_correlacion.corr(), annot=True, cmap='coolwarm', linewidths=5, vmin=0, vmax=1, cbar=False)
plt.show()
```
