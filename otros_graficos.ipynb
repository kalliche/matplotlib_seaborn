```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
data
```

## Histograma

```python
plt.hist(data, bins=10, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

## Boxplot
(grafico de cajas o bigoters)

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
data2 = data
data2 = np.append(data2,200)
plt.boxplot(data2, vert=False, )
plt.show()
```

```python
# sacar los datos atipicos
plt.boxplot(data2, vert=False, showfliers=False)
plt.show()
```

## Scatter

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n)) ** 2
color = np.random.rand(n)
```

```python
plt.scatter(x,y, s=area, c=color, alpha=0.5)
plt.show()
```
