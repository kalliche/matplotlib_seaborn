```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['colombia','india','japon','mexico','alemania']
population = [1300,800,990,1200,1000]
```

```python
plt.bar(country,population, width=0.5, color=['gray','blue'])
plt.xticks(np.arange(5),('Colombia','India','Japon','Mexico','Alemania'))
plt.show()
```

```python
plt.bar(country,population, width=0.5, color=['gray','blue'])
plt.xticks(np.arange(5),('Colombia','India','Japon','Mexico','Alemania'), rotation=90)
plt.show()
```

## reto

```python
plt.bar(country,population, width=0.5, color=['#8ecae6','#219ebc','#023046','#ffb703','#fb8500'])
plt.xticks(np.arange(5),('Colombia','India','Japon','Mexico','Alemania'))
plt.title('Population by Country')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```
