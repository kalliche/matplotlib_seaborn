```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'b')
```

```python
# dentro de la figura crea una fila pero dos columnas (dos graficos)
fig, axes = plt.subplots(nrows=1, ncols=2)  # crea un arreglo
axes[0].plot(x,y, 'b')
axes[1].plot(y,x, 'r')
```

```python
# crear una tupla con el arreglo
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)  # crea un arreglo
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
```

```python
fig, axes = (ax1, ax2) = plt.subplots(nrows=2, ncols=4)  # crea un arreglo

axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x), 'y')
axes[0,2].plot(x, np.tan(x), 'r')
axes[0,3].plot(x, np.cos(x) ** 2)
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2, ncols=4)  # crea un arreglo

ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.cos(x) ** 2)
ax5.stem(x,y)
ax6.plot(y, np.cbrt(y), 'y')
ax7.plot(y, np.absolute(y), 'r')
ax8.plot(y, np.cos(y) ** 2)
fig.tight_layout() # posiciona las graficas
```
