```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('bmh')
```

## Matlab

```python
fig, axes = plt.subplots(figsize=(8,8))

axes.plot(x, x+1, 'r--')
axes.plot(x, x+2, 'y-')
axes.plot(x, x+3, 'go-')
axes.plot(x, x+4, 'b:')
```

## Pyplot

```python
fig, axes = plt.subplots(figsize=(8,8))

axes.plot(x, x+1, color = 'green', alpha=0.5, linewidth=5, linestyle='-', marker='o')
axes.plot(x, x+2, color = 'blue', linewidth=8, linestyle='--', marker='x', markersize=10, markerfacecolor='#eeff66')
axes.plot(x, x+3, color = '#66ff66', linewidth=3, linestyle='dashed', marker='v')
axes.plot(x, x+4, color = '#eeff44', linewidth=2, linestyle=':', marker='p')
```
