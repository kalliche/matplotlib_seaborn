```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips, x='tip')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', cumulative=False, bins=20)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', cumulative=False, bins=20, hue='sex')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', cumulative=False, bins=20, hue='sex', multiple='stack')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', cumulative=False, bins=20, hue='sex', multiple='dodge')
plt.show()
```

```python
sns.kdeplot(tips, x='tip', hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde', multiple='stack')
plt.show()
```
