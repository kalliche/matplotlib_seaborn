```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2)
```

```python
fig, axes = plt.subplots(1,2, figsize=(8,5))

axes[0].plot(x,y, label='$sin(x)$')
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()    #defino el lebel y activo la leyenda

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y, label='sin(x)')
plt.title('Grafica seno')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
