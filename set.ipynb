```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show()
```

```python
sns.set(style='darkgrid', palette='dark', font_scale=0.5)
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show()
```
